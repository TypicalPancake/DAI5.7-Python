# Create 'price_label' Column

```python
# create subset first
df = df_ords_prods_merge[:1000000]
```

```python
# creat price range labels 
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
# previewing dataframe
df['price_range'].value_counts(dropna = False)
```

```python
# labeling price range labels to product
df.loc[df['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

```python
df.loc[(df['prices'] <= 15) & (df['prices'] > 5), 'price_range_loc'] = 'Mid-range product'
```

```python
df.loc[df['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
df['price_range_loc'].value_counts(dropna = False)
```

# Creating 'busiest_day' Column

```python
# previewing 
df_ords_prods_merge.describe()
```

```python
# changing 'order_dow' column name to 'order_day_of_week'
df_ords_prods_merge.rename(columns = {'order_dow' : 'orders_day_of_week'}, inplace = True)
```

```python
# check frequency 
df_ords_prods_merge['orders_day_of_week'].value_counts(dropna = False)
```

```python
# determine what is the 'high traffic', 'normal traffic', 'least traffic'.
result = []

for value in df_ords_prods_merge["orders_day_of_week"]:
  if value == 0:
    result.append("High traffic")
  elif value == 4:
    result.append("Least traffic")
  else:
    result.append("Normal traffic")
```

```python
result
```

```python
# creating column for 'busiest_day' column
df_ords_prods_merge['busiest_day'] = result
```

```python
# previewing 
df_ords_prods_merge['busiest_day'].value_counts(dropna = False)
```

Creating 2nd column ''busiest_days'

```python
# creating 'busiest_days' Column to identify 'busiest days', 'normal days', 'slowest days'
```

```python
result_2 = []

for value in df_ords_prods_merge["orders_day_of_week"]:
  if value == 0 or value == 1:
    result_2.append("Busiest days")
  elif value == 4 or value == 3:
    result_2.append("Slowest days")
  else:
    result_2.append("Normal busy")
```

```python
# previewing
result_2
```

```python
# create new column 'busiest_days' 
df_ords_prods_merge['busiest_days'] = result_2
```

```python
# previewing
df_ords_prods_merge['busiest_days'].value_counts(dropna = False)
```

# Creating 'busiest_period_of_day' Column

```python
# search the values in each row for 'order_hour_of_day' column
df_ords_prods_merge['order_hour_of_day'].value_counts()
```

Will separate into 3rds to label: "Most orders" = 10,11,14,15,13,12,16,9, ; "Average Orders" = 17,8,18,19,20,7,21,22, ; "Fewest Order" = 23,6,0,1,5,2,4,3.

```python
# creating for-loop for labeling 'most orders,'average orders,' 'fewest orders'
result_3 = []

for value in df_ords_prods_merge["order_hour_of_day"]:
  if value in [10, 11, 14, 15, 13, 12, 16, 9]:
    result_3.append("Most orders")
  elif value in [23, 6, 0, 1, 5, 2, 4, 3]:
    result_3.append("Fewest orders")
  else:
    result_3.append("Average orders")
```

```python
# previewing
result_3
```

```python
# creating new column 'busiest_period_of_day
df_ords_prods_merge['busiest_period_of_day'] = result_3
```

```python
# previewing
df_ords_prods_merge['busiest_period_of_day'].value_counts(dropna = False)
```
